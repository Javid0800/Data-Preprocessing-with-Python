17. Remove any special characters (e.g., $, ,,GB) from the Price and RAM columns and convert them to numerical format

20. Find the average price of mobile phones for each brand.

```python
exchange_rates = {
                  'Pakistan': 0.0036, # 1 PKR ≈ 0.0036 USD
                  'India': 0.012,     # 1 INR ≈ 0.012 USD
                  'China': 0.14,      # 1 CNY ≈ 0.14 USD
                  'USA': 1.0,         # 1 USD = 1 USD
                  'Dubai': 0.27       # 1 AED ≈ 0.27 USD
                   }

price_columns = [
                  'Launched Price (Pakistan)',
                  'Launched Price (India)',
                  'Launched Price (China)',
                  'Launched Price (USA)',
                  'Launched Price (Dubai)'
                ]

for column in price_columns:
  country = column.split('(')[-1].split(')')[0]
  df[column + ' (USD)'] = df[column] * exchange_rates[country]

df.head()
```

```python
df.drop(columns=['Launched Price (Pakistan)','Launched Price (India)','Launched Price (China)','Launched Price (USA)','Launched Price (Dubai)'], inplace=True)
df.head()
```

```python
price_columns_USD = ['Launched Price (Pakistan) (USD)',
                     'Launched Price (India) (USD)',
                     'Launched Price (China) (USD)',
                     'Launched Price (USA) (USD)',
                     'Launched Price (Dubai) (USD)'
                     ]

df['Average Price (USD)'] = df[price_columns_USD].mean(axis=1, skipna=True)
average_prices_by_brand = df.groupby('Manufacturer')['Average Price (USD)'].mean()
average_prices_by_brand
```

25. Find the total number of mobile phones in each Operating System category.

```python
df['Manufacturer'].unique()
```

```python
def get_operating_system(manufacturer):
  manufacturer = manufacturer.lower()
  if manufacturer == 'apple':
    return 'IOS'
  elif manufacturer in ['samsung', 'oneplus', 'vivo', 'iqoo', 'oppo',
                           'realme', 'xiaomi', 'lenova', 'motorola',
                           'huawei', 'nokia', 'sony', 'google', 'tecno',
                           'infinix', 'honor','poco']:
    return 'Android'
  else:
    return 'Unknown'
df['Operating System'] = df['Manufacturer'].apply(get_operating_system)
os_counts = df['Operating System'].value_counts()

print(os_counts)
```

26. Group by Brand and count how many phones have more than 128GB storage.

```python
df
```

```python
df['storage'] = df['Model Name'].str.extract(r'(\d+GB|\d+TB)')

def convert_to_float(storage_value):
  if 'TB' in str(storage_value):
      gb_value = int(storage_value.replace('TB','').strip())*1024
      return float(gb_value)
  elif 'GB' in str(storage_value):
      return float(storage_value.replace('GB','').strip())

df['storage'] = df['storage'].apply(convert_to_float)
```

```python
brand_memory_count = df[df['storage'] > 128].groupby('Manufacturer').size()
brand_memory_count
```
