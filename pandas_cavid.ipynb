Series

Pandasda öyrənəcəyimiz ilk əsas data tipi Seriesdir. Series-ə tək sütunlu bir cədvəl kimi baxmaq olar. Series NumPy massivinə çox bənzəyir (əslində o, NumPy massivi obyektinin üstündə qurulub). Series-i NumPy massivindən fərqləndirən cəhət onun indexlərini özümüzün təyin edə biləsidir. Yəni, yalnız ədədi indekslər deyil, istifadəçinin təyin etdiyi etiketlər (labels) ilə də indekslənə bilər.

```python
my_data = [4,5.5,6]
arr = pd.array([1,2.2,3])
```

```python
print(type(my_data))
print(type(arr))
```

```python
arr
```

```python
pd.Series([1.5,2.5]) # Series'in yaradilmasi default olaraq indeksleme ededlerle yazilir
```

```python
labels=['aa','bb','cc'] # indekslemeni oz isteyimize gorede deyise bilerik
pd.Series(data=my_data,index=labels,name="1") # name argumenti ile series'e ad veririk
```

```python
pd.Series(data = arr) #series yaratmaq ucun movcud olan massivden istifade ede bilerik
```

```python
# Lüğətin (dict) Series-ə çevrilməsi
d = {"w1":1,"w2":2,"w3":3}
pd.Series(d)
```

```python
labels=["a1","w2","w1"]
pd.Series(d,index=labels) # burda a1 indeksinin deyeri olmadigina gore NaN verdi
```

```python
sr1=pd.Series(data=[5,2,4,1],index=["Turkey","ABD","Switzerland","German"])
sr1
```

```python
sr2=pd.Series([7,6,9,8],["Switzerland","ABD","German","England"])
sr2
```

```python
sr3=sr2+sr1 # burada deyerleri toplayiriq
print(sr3,'\n')
# NaN-ləri 0 ilə əvəz etmək
sr3_filled = sr2.add(sr1, fill_value=0) #burda add() metodunnan istifade edirik
# Bu metod sr2 Series-inə sr1 Series-ini toplayır, əgər indeks uyğun gəlmirsə, həmin indeks üçün 0 dəyərini götürür.
print(sr3_filled)
```

```python
print(sr3.value_counts(),'\n') # deyerlerin saylarina baxiriq
print(sr3_filled.value_counts())
```

# Groupby

```python
data = {
        'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
        'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
        'Sales':[200,120,340,124,243,350],
        'X':[4,5,8,9,5,6]
       }
```

```python
df = pd.DataFrame(data) # lugeti (dict) dataframe'e cevirdik
df
```

```python
by_comp = df.groupby('Company').min() # Company sutunundaki her qrup ucun min tapir
by_comp
```

```python
df.groupby('Company').mean(numeric_only=True) # Company sutunundaki her qrupun ededi ortasini tapir
# numeric_only=True sadece numeric olan sutunlarda isleyeceyini bildirir
```

```python
df.groupby('Company').std(numeric_only=True) # Company sutunundaki her qrupun standart sapmasini tapir
```

```python
df.groupby('Company').count() # Company sutunundaki her qrupun sayini tapir
```

```python
df['Sales'].min() # Sales sutundaki min tapir
```

```python
by_comp.min() # bu funksiya qruplasdirilmis DataFrame uzerinde her bir qrup ucun min deyeri gosterir
```

```python
by_comp.max() # bu funksiya qruplasdirilmis DataFrame uzerinde her bir qrup ucun max deyeri gosterir
```

```python
by_comp.count() # bu funksiya qruplasdirilmis DataFrame uzerindeki sutunlarda olan deyerlerin sayini gosterir
```

```python
by_comp.describe() # ededi tipli sutunlar uzerinde statistiki gostericilere baxmaq
```

```python
by_comp.describe().T # .T veya .transpose() sutunla setirin yerini deyisir
```
