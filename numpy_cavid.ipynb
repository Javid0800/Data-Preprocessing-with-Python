```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,1.2,"1"])
print(arr)
print(type(arr)) # type obyektin tipini gosterir
print(arr.dtype) # (data type) massivdeki elemntlerin tipini gosterir
```

```python
#burda dtype parametrden istifade ederek, elementleri istediyimiz tipde yaza bilerik
arr = np.array([1,2,3,1.2,"1"], dtype=float)
print(arr)
print(type(arr)) # type obyektin tipini gosterir
print(arr.dtype) # (data type) massivdeki elemntlerin tipini teyin edir
```

```python
arr = np.array([1,2,3,1.2,"1",[1,2,3]],dtype=object) #burda elementlerin tipini object'de cevirdik, sebeb ise massivin icine yazdigimiz ferqli tipler oldugu kimi qalir
print(list(arr)) # List formatında çap edirik ki, vergüllərlə görünsün
```

```python
l=[1,"2",1.2,[1,2]] # bu listdir, massivden ferqli olaraq, elementleri oldugu tipde saxlayir
print(type(l),"\n") # type obyektin tipini gosterir
print(type(l[0]),'\n') # burda indeksi vererek elementin tipine tapiriq
for i in l: # her bir elementin tipine baxmaq ucun for dovrude yaza bilerik
  print(type(i))
```

```python
arr8 = np.eye((3),dtype=str) # bu iki funksiya kvadrat vahid matris yaradir (eye,identity)
arr9 = np.identity((9),dtype=str) #identity'de yalniz kvadrat vahid matris yaratmaq olur
# eye'in ise elave parametrleri var
print(arr8,"\n")
print(arr9)
```

```python
arr10 = np.arange(1,10,3) #(başlanğıc,son-1,addım)
arr10
```

```python
reshape_arr10 = arr10.reshape(1,3) #(setir,sutun) bu funksiya massivin formasini deyismek ucundur
print(reshape_arr10,"\n")
print("Massivin Ölçüsü:",reshape_arr10.ndim,"\n")
print('Forması:',reshape_arr10.shape,'\n')
print('Element sayı:',reshape_arr10.size)
```

```python
arr11 = np.linspace(1,10,13) #(başlanğıc,son,eded sayı) generating linearly spaced values - xətti məsafəli dəyərlər yaratmaq
d = np.diff(arr11)[0] #diff funksiyasi ededler arasindaki ferqi gosterir
print(f"Elemnlər arasındaki fərq:{d:.2f}",'\n',arr11) # burda f-string sintaksisinden istifade edirem
```
